## Data Preprocessing

Converting long categorical values in columns to numerical values wherever possible and removing columns with high number of missing values.

Since no target variables for the test dataset were given, I will only use the training dataset henceforth and evaluate based on a self-imposed test set consisting of 25% of the instances.

```python
train.isna().sum()
```

```python
# drop all columns with >50% missing values
# drop imdb_id for redundancy

train_df = train.drop(['belongs_to_collection', 'homepage', 'imdb_id'], axis=1)
```

```python
train_df.head()
```

```python
train_df.dtypes
```

```python
# unload the genres column
# use get_dummies for numerical values for each genre

def get_dictionary(s):
    try:
        d = eval(s)
    except:
        d = {}
    return d

train_df['genres_'] = train_df['genres'].apply(lambda x: sorted([d['name'] for d in get_dictionary(x)])).apply(lambda x: ','.join(x))

genres_vars = train_df.genres_.str.get_dummies(sep=',')
print(genres_vars.columns.values.tolist())
genres_vars.head()
```

```python
train_df = pd.concat([train_df, genres_vars], axis =1)
```

```python
# convert original language column to numerical by encoding


og_language = train_df.original_language.str.get_dummies()
```

```python
# keep only english column since it is the most common
# en =1 for english, en=0 for other
og_language = og_language['en']

train_df = pd.concat([train_df, og_language], axis =1)
```

```python
train_df.head()
```

```python
# convert original title to numerical by length

train_df['original_title'] = train_df['original_title'].str.len()
```

```python
# convert overview to numerical by length

train_df['overview'] = train_df['overview'].str.len()
```

```python
# convert poster path to numerical by length

train_df['poster_path'] = train_df['poster_path'].str.len()
```

```python
# convert prod companies to numerical by length

train_df['production_companies'] = train_df['production_companies'].str.len()
```

```python
# convert prod companies to numerical by length

train_df['production_countries'] = train_df['production_countries'].str.len()
```

```python
# convert spoken languages column to numerical by length

train_df['spoken_languages'] = train_df['spoken_languages'].str.len()
```

```python
# convert status to numerical by length

train_df['status'] = train_df['status'].str.len()
```

```python
# convert tagline to numerical by length

train_df['tagline'] = train_df['tagline'].str.len()
```

```python
# convert title to numerical by length

train_df['title'] = train_df['title'].str.len()
```

```python
# convert Keywords to numerical by length

train_df['Keywords'] = train_df['Keywords'].str.len()
```

```python
# convert cast to numerical by length

train_df['cast'] = train_df['cast'].str.len()
```

```python
# convert crew to numerical by length

train_df['crew'] = train_df['crew'].str.len()
```

```python
# convert release date to month, day, year

train_df['year'] = train_df.release_date.apply(lambda x: int(x.split('/')[2]))
train_df['month'] = train_df.release_date.apply(lambda x: int(x.split('/')[0]))
train_df['day'] = train_df.release_date.apply(lambda x: int(x.split('/')[1]))

```

```python
train_df = train_df.drop(['genres'], axis=1)
```

```python
train_df.dtypes
```

```python
# drop last object type columns

train_df = train_df.drop(['genres_', 'release_date', 'original_language'], axis=1)
```

```python
# fill missing values with mean

train_df = train_df.fillna(train_df.mean())
```

```python
train_df.head()
```

## Linear Regression

This is a regression problem because we are trying to predict revenue exactly, not classify by revenue ranges. Using linear regression should give us a numerical prediction for revenue based on the features given.

```python
from sklearn.model_selection import train_test_split
```

```python
y
```

```python
X = train_df.drop(['revenue'], axis=1)
```

```python
# use train_test_split to divide dataset into train and test sets

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# convert dataframes to numpy arrays

X_train = X_train.to_numpy()
X_test = X_test.to_numpy()
y_train = y_train.array.to_numpy()
y_test = y_test.array.to_numpy()
```

```python
# add dummy variable to each instance for X's

X_train_b = np.c_[np.ones((len(X_train), 1)), X_train]
X_test_b = np.c_[np.ones((len(X_test), 1)), X_test]
```

```python
# find weights

inv = np.linalg.inv

weights_best = inv(X_train_b.T.dot(X_train_b)).dot(X_train_b.T).dot(y_train)
```

```python
weights_best
```

```python
# predict y variables for test set using weights from train set

y_pred_test = X_test_b.dot(weights_best)
```

## Evaluate Model

We'll use the root mean squared error (RMSE) and mean square error (MSE) to evaluate the model

```python
y_test_scaled = y_test.reshape(-1,1)
y_pred_scaled = y_pred_test.reshape(-1,1)
```

```python
# scale predicted values
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

y_test_scaled = scaler.fit_transform(y_test.reshape(-1,1))
y_pred_scaled = scaler.fit_transform(y_pred_test.reshape(-1,1))
```

```python
y_pred_scaled.min()
```

```python
y_pred_scaled.max()
```

```python
# calculate using numpy

MSE = np.square(np.subtract(y_test_scaled,y_pred_scaled)).mean() 
 
RMSE = np.sqrt(MSE)
```
